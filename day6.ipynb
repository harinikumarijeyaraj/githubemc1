```python
a= int (input("enter an number: "))
o= input("enter operator: ")
b= int (input("enter a number: "))

if o=="+":
    print ("addition of number: ",(a+b))
elif o=="-":
    print ("subtraction of number: ",(a-b) )
elif o=="*":
    print ("multiple of number: ",(a*b))
elif o=="/":
    print("division of number: ",(a/b)) 
else :
     print("math error: ")

```
