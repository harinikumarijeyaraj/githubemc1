```python
text = """Python is a high-level interpreted programming language celebrated for its simple English like syntax that prioritizes readability and developer efficiency
    Since its creation by Guido van Rossum it has become a global favorite 
    because it allows complex ideas to be expressed in fewer lines of code than languages like C++ or Java
    Its batteries included philosophy means it comes with a vast standard library 
    supporting everything from web development and automation to cutting-edge Data Science and Artificial Intelligence"""

words = text.lower().split()
counts = {}
for w in words:
    counts[w] = counts.get(w, 0) + 1
print(counts)
```

```python
text = """Python is a high-level interpreted programming language celebrated for its simple English like syntax that prioritizes readability and developer efficiency
Since its creation by Guido van Rossum it has become a global favorite 
because it allows complex ideas to be expressed in fewer lines of code than languages like C++ or Java
Its batteries included philosophy means it comes with a vast standard library 
supporting everything from web development and automation to cutting-edge Data Science and Artificial Intelligence"""

words = text.lower().split()
count = {}
for w in words:
    count[w] = count.get(w, 0) + 1
for i,j in count.items():
    print(i,"-",j)
```
