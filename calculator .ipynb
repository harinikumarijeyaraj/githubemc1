```python
def calculator():
    try:
        a = float(input("Enter first number: "))
        op = input("Enter operator (+, -, *, /): ")
        b = float(input("Enter second number: "))

        if op == "+":
            print("Result:", a + b)

        elif op == "-":
            print("Result:", a - b)

        elif op == "*":
            print("Result:", a * b)

        elif op == "/":
            print("Result:", a / b)

        else:
            print("Invalid Operator")

    except ValueError:
        print(" Enter valid numbers")

    except ZeroDivisionError:
        print(" Cannot divide by zero")
        
calculator()
```
