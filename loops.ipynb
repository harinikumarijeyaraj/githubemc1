```python
l = [1,2,1,2,3,3,2,5,4,6,4]
k = []
for i in l:
    if i not in k:
        print(i, "-", k)
        k.append(i)
```

```python
l=[1,2,3,4,5,6,1,2,3,4]
k=[]
for i in l:
    if i not in k:
        a=l.count(i)
        print(i,"-",a)
        k.append(i)
```
